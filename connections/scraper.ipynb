```python
def get_data(puzzle_number, word_map = {}):
    url = f"https://connections.swellgarfo.com/nyt/{puzzle_number}"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser", from_encoding="utf-8")
    answer = soup.find(id="__NEXT_DATA__").contents
    answer = answer[0]
    answer = json.loads(answer)
    answer = answer['props']['pageProps']['answers']
    for i in answer:
        for word in i['words']:
            if word in word_map:
                word_map[word].append(i['description'])
            else:
                word_map[word] = [i['description']]

    return word_map
```

```python
word_map = {}
for i in range(1, 263):
    word_map = get_data(i, word_map)
word_map
```

```python
with open('word_map.json', 'w') as f:
    json.dump(word_map, f, ensure_ascii=False, indent=4)
```

```python
reversed_dict = {}
for key, values in word_map.items():

    for value in values:

        if value in reversed_dict:

            reversed_dict[value].append(key)

        else:

            reversed_dict[value] = [key]

reversed_dict
```

```python
with open('reversed_word_map.json', 'w') as f:
    json.dump(reversed_dict, f, ensure_ascii=False, indent=4)
    
    
```

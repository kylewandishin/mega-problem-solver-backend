```python
def find_similar_words(guess, semanticscore):
    arr = {}
    # for word in list(model.key_to_index):
    for word in words:
        similarity = abs((np.dot(model[guess], model[word])/(np.linalg.norm(model[guess])* np.linalg.norm(model[word])) * 100) - semanticscore)
        
        if similarity <= 0.001:
            arr[word] = similarity

    return arr
            
```

```python
def square_rooted(x):
    return round(np.sqrt(sum([a*a for a in x])),3)

def cosine_similarity(x,y):
    numerator = sum(a*b for a,b in zip(x,y))
    denominator = square_rooted(x)*square_rooted(y)
    ans = round(numerator/float(denominator),4)
    return (ans * 100)
```

```python
def find_similar_words(word, sem_score):
    ans = []
    for w in words:
        if (abs((cosine_similarity(model[word], model[w])-sem_score)) <= 0.1):
            ans.append(w)
    return ans
```
